```
path = '/content/drive/Shareddrives/dacon/404/preprocessed/'
trainset = pd.read_csv(path + 'stack_train.csv', index_col=['Unnamed: 0'])
testset = pd.read_csv(path + 'stack_test.csv', index_col=['Unnamed: 0'])
```

```
from sklearn.preprocessing import RobustScaler
```

```
scaler = RobustScaler()
```

```
trainset.loc[:,'LGBM_0' : 'GBC_1'] = scaler.fit_transform(trainset.loc[:,'LGBM_0' : 'GBC_1'] )
```

```
testset.loc[:,'LGBM_0' : 'GBC_1'] = scaler.transform(testset.loc[:,'LGBM_0' : 'GBC_1'] )
```

```
clf = setup(data = trainset, target = 'TARGET', session_id = 42) 
```

```
lgbm = create_model('lightgbm')
```

```
tune_lgbm = tune_model(lgbm)
```

```
predictions = predict_model(tune_lgbm, data = testset)
```

```
x = []
for i in range(len(predictions['Score'])):
  if predictions['Label'][i+30000] ==1:
    x.append(predictions['Score'][i+30000])
  else:
    x.append(1-predictions['Score'][i+30000])
```

```
sample_submssion = pd.read_csv('/content/drive/Shareddrives/dacon/404/sample_submission.csv')
sample_submssion['problem'] = x
sample_submssion.to_csv("LGBM_stack_caret.csv", index = False)
!cp LGBM_stack_caret.csv "drive/My Drive/"
```
