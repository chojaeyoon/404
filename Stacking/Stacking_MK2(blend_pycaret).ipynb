```
path = '/content/drive/Shareddrives/dacon/404/preprocessed/'
trainset = pd.read_csv(path + 'stack_train.csv', index_col=['Unnamed: 0'])
testset = pd.read_csv(path + 'stack_test.csv', index_col=['Unnamed: 0'])
```

```
clf = setup(data = trainset, target = 'TARGET', session_id = 42) 
```

```
best_5 = compare_models(sort = 'AUC', n_select = 3)
```

```
blended = blend_models(estimator_list = best_5, fold = 10, method = 'soft')
pred_holdout = predict_model(blended)
final_model = finalize_model(blended)
```

```
predictions = predict_model(final_model, data = testset)
```

```
x = []
for i in range(len(predictions['Score'])):
  if predictions['Label'][i+30000] ==1:
    x.append(predictions['Score'][i+30000])
  else:
    x.append(1-predictions['Score'][i+30000])
```
