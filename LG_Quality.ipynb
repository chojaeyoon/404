# FWVER

```python
# fwver의 임계값 먼저 찾기
# fwver는 quaility, error 데이터 모두에 존재하기 때문에, 두개의 데이터 모두 train,test 데이터 합쳐서 임계값을 정함.

qui2_df = pd.concat([train_qui2, test_qui2], axis = 0)
err2_df = pd.concat([train_err2, test_err2], axis = 0)
fwver_df = pd.concat([qui2_df, err2_df], axis = 0)
fwver_df
```

```python
# fwver에 어떤 값들이 들어있는지 확인 - nan 제외 47개

fwver_df['fwver'].unique()

# 문제가 되는 것은 8.5.3, 10 두개를 변환해주어야 함.
# len(fwver_df['fwver'].unique())
```

```python
# 형식 통일 

fwver_df.replace({'fwver': np.nan}, {'fwver': '00.00.0000'}, inplace = True)
fwver_df.replace({'fwver': '8.5.3'}, {'fwver': '08.05.3000'}, inplace = True)
fwver_df.replace({'fwver': '10'}, {'fwver': '10.00.0000'}, inplace = True)
```

```python
fwver_df['fwver'].unique()
```

```python
# 범주형 변수 -> 연속형 변수로 만드는 방법은 많은데, information loss를 감수하고 연속형 변수 -> 범주형 변수를 하는 경우가 적어 자료가 적음.
# '통계적으로 유의미'하게 범주를 나누는 방법은 없고, 가장 현실적인 게 범주의 개수를 설정하고 그 안의 개수를 동일하게 해주는 방법임.

fwver_df2 = fwver_df.copy()
fwver_df2['fwver'] = fwver_df2['fwver'].str.replace('\.','', regex = True)
fwver_df2
```

```python
# fwver type 확인
print(fwver_df2.dtypes)
```

```python
fwver_df2 = fwver_df2.astype({'fwver': 'int'})
fwver_df2
```

```python
def jitter(a_series, noise_reduction=1000000):
    return (np.random.random(len(a_series))*a_series.std()/noise_reduction)-(a_series.std()/(2*noise_reduction))

fmver_cat1 = pd.qcut(fwver_df2.fwver + jitter(fwver_df2.fwver), 10, labels=False) # 숫자 바꿀 수 있음.
fmver_cat2 = pd.qcut(fwver_df2.fwver, 10, labels=False, duplicates = 'drop')
```

```python
# 요약정리
# qcut이 하나의 값은 같은 카테고리로 들어가야한다게 있어서 이를 무시하는 jitter함수를 통해 해결함.
# grouped1은 같은 값도 여러 개의 카테고리로 나누는 함수
# grouped2는 같은 값은 무조건 하나의 카테고리
# grouped2 기준으로 해서 원하는 카테고리수로 하면 되지 않나 생각.
# 10개를 설정해도 8개로 분류되는 것을 보면 8개가 최대인듯 - 8개로 해서 분류하는게 최선.

def summary_func(group):
    return {'count': group.count(), 
             'mean': group.mean(), 
             'std': group.std(), 
             'min': group.min(), 
             'max': group.max()}

grouped1 = fwver_df2.fwver.groupby(fmver_cat1)
grouped2 = fwver_df2.fwver.groupby(fmver_cat2)

#grouped1.apply(summary_func).unstack()
grouped2.apply(summary_func).unstack()
```
